```r
head(protein_df)
```

```r
# adding a col to denote whether a protein had increased or decreased expression

# add a column of NAs
protein_df$diffexpressed <- "No Significance"

# if log2FC > 0.6 and P.Value < 0.05, set as "UP" 
protein_df$diffexpressed[protein_df$log2FC > 0.6 & protein_df$P.Value < 0.05] <- "Increased Significantly"

# if log2FC < -0.6 and P.Value < 0.05, set as "DOWN"
protein_df$diffexpressed[protein_df$log2FC < -0.6 & protein_df$P.Value < 0.05] <- "Decreased Significantly"

protein_df$diffexpressed = factor(protein_df$diffexpressed, levels = c("Decreased Significantly", "Increased Significantly", "No Significance"))
```

```r
options(repr.plot.width=13, repr.plot.height=6) #changing size

#choosing palette from package
metbrewer_colors = MetPalettes$Tam[[1]]

ggplot() + 
  geom_point(data = protein_df, aes(x = log2FC, y = -log10(P.Value), color = diffexpressed), size = 3) + 
  #geom_hline(data = protein_df, aes(yintercept = -log10(0.05)), linetype = "dotdash") + 

  #labeling points with statistical significance and FC
  geom_text_repel(data = protein_df %>%
                      filter(abs(log2FC) > 0.6 & P.Value < 0.05 & diffexpressed == "Increased Significantly"), 
    aes(x = log2FC, y = -log10(P.Value), label = Protein, color = diffexpressed),
    xlim = c(2, NA),
    hjust = 0,
    segment.size = 0.4,
     box.padding = 0.7,
     max.overlaps = Inf,
    size = 5, 
    show.legend = FALSE) + 

  geom_text_repel(data = protein_df %>%
                      filter(abs(log2FC) > 0.6 & P.Value < 0.05 & diffexpressed == "Decreased Significantly"), 
    aes(x = log2FC, y = -log10(P.Value), label = Protein, color = diffexpressed),
    xlim = c(NA, -2),
    hjust = 1,
    segment.size = 0.4,
     box.padding = 0.7,
     max.overlaps = Inf,
    size = 5, 
    show.legend = FALSE) + 

  facet_wrap(~Treatment) + 

  theme_light() + 
  theme(axis.line = element_line(colour = "black"), #making x and y axes black
        axis.text = element_text(size = 12), #changing size of x axis labels
        axis.title = element_text(face = "bold", size = rel(1.7)), #changes axis titles
        legend.title = element_text(face = 'bold', size = 14), #changes legend title
        legend.text = element_text(size = 12), #changes legend text
        strip.text.x = element_text(size = 15, face = "bold"), #changes size of facet x axis 
        strip.text.y = element_text(size = 15, face = "bold")) + #changes size of facet y axis 
  labs(x = 'Fold Change (log2)', y = 'P Value (-log)') + 

  scale_color_manual(name = "Expression",
                      values = c(metbrewer_colors[2], metbrewer_colors[5], metbrewer_colors[7]))
```
